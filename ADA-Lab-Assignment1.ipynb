```python
#T1
import time
import matplotlib.pyplot as plt

# 1. Constant Time: O(1)
def constant_algo(n):
    return 1  # Always one step

# 2. Linear Time: O(n)
def linear_algo(n):
    count = 0
    for i in range(n):
        count += 1
    return count

# 3. Quadratic Time: O(n^2)
def quadratic_algo(n):
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count

# 4. Logarithmic Time: O(log n)
def logarithmic_algo(n):
    count = 0
    while n > 1:
        n //= 2
        count += 1
    return count

input_sizes = [10, 100, 500, 1000]
results = {"O(1)": [], "O(n)": [], "O(n^2)": [], "O(log n)": []}
```

```python
# Measuring Time
for n in input_sizes:
    # O(1)
    start = time.perf_counter()
    constant_algo(n)
    results["O(1)"].append(time.perf_counter() - start)

    # O(n)
    start = time.perf_counter()
    linear_algo(n)
    results["O(n)"].append(time.perf_counter() - start)

    # O(n^2)
    start = time.perf_counter()
    quadratic_algo(n)
    results["O(n^2)"].append(time.perf_counter() - start)

    # O(log n)
    start = time.perf_counter()
    logarithmic_algo(n)
    results["O(log n)"].append(time.perf_counter() - start)
```

```python
plt.figure(figsize=(8, 5))
for label, times in results.items():
    plt.plot(input_sizes, times, label=label, marker='o')

plt.title("Task 1: Input Size vs Execution Time")
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#T2

import time
import random
import matplotlib.pyplot as plt

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

sizes = [100, 1000, 5000, 10000]
ls_worst_times = []
bs_worst_times = []

for n in sizes:
    arr = sorted(random.sample(range(n * 10), n))
    target = -1

    start = time.perf_counter()
    linear_search(arr, target)
    ls_worst_times.append(time.perf_counter() - start)

    start = time.perf_counter()
    binary_search(arr, target)
    bs_worst_times.append(time.perf_counter() - start)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(sizes, ls_worst_times, label="Linear Search (Worst Case)", marker='s')
plt.plot(sizes, bs_worst_times, label="Binary Search (Worst Case)", marker='^')
plt.title("Task 2: Search Performance Comparison")
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# T3
factorial_calls = 0

def factorial(n):
    global factorial_calls
    factorial_calls += 1

    if n == 0:
        return 1
    return n * factorial(n - 1)

```

```python
fib_recursive_calls = 0

def fib_recursive(n):
    global fib_recursive_calls
    fib_recursive_calls += 1

    if n <= 1:
        return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)

```

```python
def fib_dp(n):
    if n <= 1:
        return n

    dp = [0] * (n + 1)
    dp[1] = 1

    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]

    return dp[n]
```

```python
n_values = [5, 10, 20]

print("n | Factorial Calls | Fact Time | Fib Rec Calls | Fib Rec Time | Fib DP Time")
print("-"*80)

for n in n_values:
    # Factorial
    factorial_calls = 0
    start = time.time()
    factorial(n)
    fact_time = time.time() - start

    # Fibonacci Recursive
    fib_recursive_calls = 0
    start = time.time()
    fib_recursive(n)
    fib_rec_time = time.time() - start

    # Fibonacci DP
    start = time.time()
    fib_dp(n)
    fib_dp_time = time.time() - start

    print(f"{n} | {factorial_calls:^15} | {fact_time:.6f} | "
          f"{fib_recursive_calls:^14} | {fib_rec_time:.6f} | {fib_dp_time:.6f}")
```

```python
# T4
import time

calls_rec1 = 0

def recurrence_1(n):
    global calls_rec1
    calls_rec1 += 1

    if n <= 1:
        return n
    return recurrence_1(n // 2) + n
```

```python
calls_rec2 = 0

def recurrence_2(n):
    global calls_rec2
    calls_rec2 += 1

    if n <= 1:
        return n
    return recurrence_2(n // 2) + recurrence_2(n // 2) + n
```

```python
n_values = [10, 100, 500, 1000]

print("n | Rec1 Calls | Rec1 Time | Rec2 Calls | Rec2 Time")
print("-" * 60)

for n in n_values:
    # Recurrence 1
    calls_rec1 = 0
    start = time.time()
    recurrence_1(n)
    time_rec1 = time.time() - start

    # Recurrence 2
    calls_rec2 = 0
    start = time.time()
    recurrence_2(n)
    time_rec2 = time.time() - start

    print(f"{n:<4}| {calls_rec1:^10} | {time_rec1:.6f} | "
          f"{calls_rec2:^10} | {time_rec2:.6f}")
```
